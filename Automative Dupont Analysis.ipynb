```r
#starts from here
getfinstat <- function(link){
    data <- read.csv2(link, sep=",",  header=FALSE,blank.lines.skip = FALSE,fill = TRUE,na.strings = "NA",stringsAsFactors = FALSE)
    colnames <- c('category')
    for (i in 2:(ncol(data)))
    {
    colnames[i] <- data[2,i]
    }
    df <- data.frame(data[3:nrow(data),1:ncol(data)],row.names = c(1:(nrow(data)-2)))
    colnames(df) <- colnames
    return(df)
}
```

```r
getfinstat(BalanceS)
```
